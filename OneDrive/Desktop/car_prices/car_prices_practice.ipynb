```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('cardekho_imputated.csv')
df.head()
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
import matplotlib.pyplot as plt
```

```python
df['car_name'].value_counts()
```

```python
df['car_name'].nunique()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(drop='first',sparse_output=False) 
```

```python
encoded_features=encoder.fit_transform(df[['car_name','fuel_type','transmission_type','seller_type']])
encoded_features.shape
```

```python
encoded_df = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out(['car_name','fuel_type','transmission_type','seller_type'],))
```

```python
encoded_df=encoded_df.astype(int)
```

```python
encoded_df.head()
```

```python
final_df = pd.concat([df.drop(['car_name','fuel_type','transmission_type','seller_type'], axis=1), encoded_df], axis=1)
```

```python
final_df.head()
```

```python
final_df.drop(['brand','model'],axis=1,inplace=True)
final_df.head()
```

```python
tree_model=DecisionTreeRegressor()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=final_df.drop('selling_price',axis=1)
y=final_df[['selling_price']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
tree_model.fit(x_train,y_train)
```

```python
y_pred=tree_model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(y_test,y_pred))
```

```python
y_test
```

```python
y_pred=pd.DataFrame(y_pred)
y_pred
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'max_depth':[None,10,20,30,40],
           'min_samples_leaf':[1,5,10],
           'min_samples_split':[2,10,20],
           'max_features':[None,'sqrt','log2'],
           'max_leaf_nodes':[None,10,20,30,40]}
```

```python
grid_search=GridSearchCV(estimator=tree_model,param_grid=param_grid,cv=5,scoring='neg_mean_squared_error',n_jobs=-1)
```

```python
grid_search.fit(x_train,y_train)
```

```python
y_pred_new=grid_search.predict(x_test)
```

```python
print(r2_score(y_test,y_pred_new))
```
